## Questions

With your own data that you collected:
1. Create 2 graphs between two variables. Be sure to include the title, labels of both axis. Include a legend if necessary. Be sure that nothing overlaps in the graph (words or legends, etc). 

2. Write 1-2 sentences that describe the main information presented and a conclusion that you could draw from that graph. 

3. Display the graphs side by side in one figure. 

4. Do a third graph that is not a line graph or bar chart (with proper labels and titles). Make sure this third graph displays different information than presented in step 1.


```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Re-creating dataset from Assignment 2
# Source: https://restcountries.com/
url = 'https://restcountries.com/v3.1/all?fields=name,population,area,region,subregion,languages,borders'
response = requests.get(url)
countries_raw = response.json()

records = []
for country in countries_raw:
    name          = country.get('name', {}).get('common', None)
    population    = country.get('population', None)
    area          = country.get('area', None)
    region        = country.get('region', None)
    languages     = country.get('languages', {})
    num_languages = len(languages) if languages else 0
    borders       = country.get('borders', [])
    num_borders   = len(borders) if borders else 0
    records.append({'country': name, 'population': population, 'area_km2': area,
                    'region': region, 'num_languages': num_languages, 'num_borders': num_borders})

df = pd.DataFrame(records)
df = df.dropna(subset=['population', 'area_km2'])
df = df[(df['area_km2'] > 0) & (df['population'] > 0)].reset_index(drop=True)
df['population_density'] = (df['population'] / df['area_km2']).round(2)
df.head()
```

```python
# Graph 1: Average population density by region
# Source: https://r02b.github.io/seaborn_palettes/
avg_density = df.groupby('region')['population_density'].mean().reset_index().sort_values('population_density', ascending=False)

sns.set_style("whitegrid")
plt.figure(figsize=(9, 5))
plt.bar(avg_density['region'], avg_density['population_density'],
        color=sns.color_palette('muted', len(avg_density)), edgecolor='white')
plt.xlabel('World Region', fontweight='bold')
plt.ylabel('Avg. Population Density (people per km²)', fontweight='bold')
plt.title('Average Population Density by World Region', fontsize=14, fontweight='bold')
plt.xticks(rotation=20, ha='right')
plt.tight_layout()
plt.show()

# Asia has the highest average population density, while Oceania and Antarctica have the lowest.
# This suggests historical and geographic factors lead to very uneven distributions of people across regions.
```

```python
# Graph 2: Number of borders vs population density, colored by region
# Removing top 1% density outliers (city-states like Monaco) for readability
df_filtered = df[df['population_density'] <= df['population_density'].quantile(0.99)].copy()

plt.figure(figsize=(10, 6))
ax = sns.scatterplot(x='num_borders', y='population_density', data=df_filtered,
                     hue='region', palette='muted', marker='o', alpha=0.7)
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Region', fontsize=9)
# legend placed outside so nothing overlaps — Lecture 8 & 9
plt.xlabel('Number of Bordering Countries', fontweight='bold')
plt.ylabel('Population Density (people per km²)', fontweight='bold')
plt.title('Number of Borders vs. Population Density by Region', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# There is no strong relationship between number of borders and population density.
# Island nations with 0 borders span a wide range of densities, while landlocked countries cluster at moderate levels.
```

```python
# Displaying both graphs side by side — Lecture 9
# Source: https://jakevdp.github.io/PythonDataScienceHandbook/04.08-multiple-subplots.html
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

ax1 = plt.subplot(1, 2, 1)
ax1.bar(avg_density['region'], avg_density['population_density'],
        color=sns.color_palette('muted', len(avg_density)), edgecolor='white')
ax1.set_xlabel('World Region', fontweight='bold')
ax1.set_ylabel('Avg. Population Density (people / km²)', fontweight='bold')
ax1.set_title('Avg. Population Density by Region', fontweight='bold')
ax1.set_xticks(range(len(avg_density['region'])))
ax1.set_xticklabels(avg_density['region'], rotation=20, ha='right')

ax2 = plt.subplot(1, 2, 2)
for i, r in enumerate(df_filtered['region'].unique()):
    d = df_filtered[df_filtered['region'] == r]
    ax2.plot(d['num_borders'], d['population_density'], linestyle='none',
             marker='o', color=sns.color_palette('muted', 8)[i], label=r, alpha=0.7)
ax2.set_xlabel('Number of Bordering Countries', fontweight='bold')
ax2.set_ylabel('Population Density (people per km²)', fontweight='bold')
ax2.set_title('Borders vs. Population Density', fontweight='bold')
ax2.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Region', fontsize=8)

fig.suptitle('Country Population Density — Two Views', fontsize=16, fontweight='bold')
fig.tight_layout()
plt.show()
```

```python
# Graph 3: Box plot of number of languages by region (not a line or bar chart!)
# Shows linguistic diversity — completely different info from the density graphs above
df_box = df[df['region'] != 'Antarctic'].copy()
region_order = df_box.groupby('region')['num_languages'].median().sort_values(ascending=False).index.tolist()

sns.set_style("white")
plt.figure(figsize=(11, 6))
ax = sns.boxplot(x='region', y='num_languages', data=df_box, order=region_order,
                 hue='region', palette='muted', legend=False)
ax.set_xlabel('World Region', fontweight='bold')
ax.set_ylabel('Number of Official Languages', fontweight='bold')
ax.set_title('Distribution of Official Languages per Country by Region', fontsize=14, fontweight='bold')
ax.set_xticks(range(len(region_order)))
ax.set_xticks(range(len(region_order)))
ax.set_xticklabels(region_order, rotation=15, ha='right')
plt.tight_layout()
plt.show()

# Africa has the highest median number of official languages, showing its rich linguistic diversity.
# This is different from the density graphs — it's about cultural composition, not population size.
```
